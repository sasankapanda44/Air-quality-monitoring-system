```python
from sklearn.svm import SVR
```

```python
X  = np.vstack ([df1_1[['SO2','NH3','O3','PM 2_5','PM_10']].values ,df2_1[['SO2','NH3','O3','PM2.5','PM10']].values])
```

```python
y1 = df1_1['Avg_AQI'].values
y1 = y1.reshape(-1,1)
y2 = df2_1['AQI'].values
y2 = y2.reshape(-1,1)

y= np.vstack ([y1,y2])
```

```python
model = SVR()

model.fit(X,y)
```

```python
from sklearn.metrics import r2_score,mean_squared_error

print(r2_score(y,model.predict(X)))
print(mean_squared_error(y,model.predict(X))**0.5)
```
